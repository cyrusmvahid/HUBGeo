# Data Types
- If we add two apples to three oranges, what will we have? two apples to three oranges
- If we add two apples to three apples, what will we have? 5 apples
- If we add two apples and three oranges, how many individual fruits will we have? 5
- We can apply operators only to variables of the same size.
- In programming variables are simply the type of memory that holds values and are not concerned with semantic compatibility.

## String
- String is an enumerable list of single characters.

````python
fruit1 = 'Apple'
fruit2 = 'Orange'
print("{} + {} = {}".format(fruit1, fruit2, fruit1+fruit2))
print("type of \"{}\" is {} and has a length of {}.".format(fruit1, type(fruit1), len(fruit1)))
````

## Integer
- Whole umbers or numbers without the fractional parts.

````python
a, b = 5, 3
print("{} + {} = {}\n{} - {} = {}\n{} * {} = {}\n{} \ {} = {}".format(a, b, a+b,
                                                                      a, b, a-b,
                                                                      a, b, a*b, 
                                                                      a, b, a/b))
print("type of \"{}\",\"{}\", \"{}\", and \"{}\" respectively are {}, {}, {}, {}:".format(a, b, a+b, a/b,
                                                                                          type(a), 
                                                                                          type(b), 
                                                                                          type(a+b), 
                                                                                          type(a/b),))
````

## Float
-  float values are specified with a decimal point.

````python
a, b = 5.3, 3.0
print("{} + {} = {}\n{} - {} = {}\n{} * {} = {:4.2f}\n{} \ {} = {:4.2f}".format(a, b, a+b,
                                                                      a, b, a-b,
                                                                      a, b, a*b, 
                                                                      a, b, a/b))
print("type of \"{}\",\"{}\", \"{}\", and \"{:4.2f}\" respectively are {}, {}, {}, {}:".format(a, b, a+b, a/b,
                                                                                          type(a), 
                                                                                          type(b), 
                                                                                          type(a+b), 
                                                                                          type(a/b)))
````

##  Char
-  A string of length 1 that can also be represented as an integer.

````python
a, b,= 'a', 'b'
c = a+b
print("a={}; b={}; c={}".format(a, b, c))
print("type of a, b, and c are {}, {}, {} respectively".format(type(a), type(b), type(c)))
print("length of a, b, and c are {}, {}, {} respectively".format(len(a), len(b), len(c)))
print("int representatino of \'a\' and \'b\' are {} and {} respectively".format(ord(a), ord(b)))
print("char representatino of 97 and 98 are {} and {} respectively".format(chr(97), chr(98)))
````

## Dictionaries
- Dictionaries are similar to lists, but instead of holding and index, they are forms as key-value pairs.

````python
d1 = {'a':'element1',  
      'b':'element2', 
      'c': 'element3', 
      'd': 1953}
d1['a'] = 'element1 is dead'
d1['new_item']='new item'
print("d1: {}\nd1['a']: {}\ntype(d1): {}\ntype(d1['a']: {}\ntype(d1['d']: {}))".format(d1, 
                                                                                       d1['a'], 
                                                                                       type(d1), 
                                                                                       type(d1['a']), 
                                                                                       type(d1['d'])))
````

### Basic Operations

````python
(d1.values(), d1.keys(), 'new_item' in d1, 'new item' in d1, d1.fromkeys(['a', 'b']))
````

````python
d1.pop('new_item')
d2 = dict([('new_item', 'new item')])
d3=d1.copy()
d3.update(d2)
(d1, d2, d3)
````
